```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()  

x = diabetes.data[:, 2]
y = diabetes.target

w = 1.0
b = 1.0

# Update a weight and an intercept using 'backpropagation'

# Now, we will multiply the error(between y^ and y) to the rate of change of y^.
# So, if a gap of y^ and y is huge, w and b will be updated faster.
# Moreover, y^hat is bigger than y, the error will be negative,
# So, we also can update w and b in the opposite way(to decrease y^) automatically.
```

The rate of change of y^ in terms of w is the same as the feature value.
<br>
The rate of change of y^ in terms of b is the same as 1. (always)

```python
for x_i, y_i in zip(x, y): # take out each x and y from the input set and target set orderly.
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate  = x_i  # rate of change of y^ in terms of w is the same as the feature value
    w = w + w_rate * err 
    b = b + 1 * err # rate of change of y^ in terms of b is the same as 1 (always)

    
print(w, b)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
pt1 = (-0.1, -0.1*w + b)
pt2 = (0.15, 0.15*w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
for i in range(100):
    for x_i, y_i in zip(x, y): 
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate  = x_i 
        w = w + w_rate * err 
        b = b + 1 * err 
    
print(w, b)

# Now we have a decent model that can describe the relationship between the second feature data and target data.
# Model: y_hat = 913.6x + 123.4
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
pt1 = (-0.1, -0.1*w + b)
pt2 = (0.15, 0.15*w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Let's predict the target value when we assign a random feature value.

x_new = 0.18
y_pred = x_new * w + b
print(y_pred)
```

```python
plt.scatter(x, y)
plt.scatter(x_new, y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Orange dot is predicted point.
```
